```python
# library yang diperlukan
import pandas as pd

# Promt : beri kode untuk membaca dataset Car_sales.csv
# Baca dataset Car_sales.csv
df = pd.read_csv('/content/drive/My Drive/Machine Learning/Car_sales.csv')
print(df.head())
```

```python
# Promt : beri kode untuk menampilkan boxplot dari fitur numerik masing masing memiliki boxplotnya tersendiri
# library yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya kolom numerik
numeric_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']
numeric_data = df[numeric_columns]

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(14, 10))
sns.boxplot(data=numeric_data, orient="h", palette="Set2", saturation=0.75)
plt.title("Boxplot of Numeric Features", fontsize=16)
plt.xlabel("Value", fontsize=14)
plt.ylabel("Feature", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Promt : beri kode untuk menampilkan grafik distribusi dari fitur numerik masing masing memiliki grafik tersendiri
# library yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya kolom numerik
numeric_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']
numeric_data = df[numeric_columns]

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat distribusi grafik untuk setiap fitur numerik
plt.figure(figsize=(14, 10))
for i, col in enumerate(numeric_data.columns):
    plt.subplot(4, 3, i + 1)
    sns.histplot(numeric_data[col], kde=True)
    plt.title(col, fontsize=12)
    plt.xlabel("")
    plt.ylabel("Frequency", fontsize=10)
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Promt : beri kode untuk menampilkan grafik violin dari fitur numerik masing masing memiliki grafik tersendiri
# library yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya kolom numerik
numeric_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']
numeric_data = df[numeric_columns]

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat grafik violin untuk setiap fitur numerik
plt.figure(figsize=(14, 10))
for i, col in enumerate(numeric_data.columns):
    plt.subplot(4, 3, i + 1)
    sns.violinplot(y=numeric_data[col], color="skyblue")
    plt.title(col, fontsize=12)
    plt.xlabel("")
    plt.ylabel("")
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# Promt : buat kode untuk membuat scater plot untuk semua fitur numerik
# library yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya kolom numerik
numeric_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']
numeric_data = df[numeric_columns]

# Mengatur gaya seaborn
sns.set(style="whitegrid")

# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(numeric_data)
plt.show()
```

```python
# Promt : buat heat map dari kolerasi fitur numerik
# library yang diperlukan
import seaborn as sns
import matplotlib.pyplot as plt

# Memilih hanya kolom numerik
numeric_columns = ['Sales_in_thousands', '__year_resale_value', 'Price_in_thousands', 'Engine_size',
                   'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity',
                   'Fuel_efficiency', 'Power_perf_factor']
numeric_data = df[numeric_columns]

# Menghitung matriks korelasi
correlation_matrix = numeric_data.corr()

# Mengatur ukuran gambar
plt.figure(figsize=(12, 10))

# Membuat heatmap korelasi
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Heatmap of Numeric Features")
plt.show()
```
