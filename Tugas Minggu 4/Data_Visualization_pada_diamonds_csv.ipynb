```python
# library yang diperlukan
import pandas as pd

# Promt : beri kode untuk membaca dataset diamonds.csv
# Baca dataset diamonds.csv
df = pd.read_csv('/content/drive/My Drive/Machine Learning/diamonds.csv')
print(df.head())
```

```python
# Promt : beri kode untuk menampilkan boxplot dari fitur numerik masing masing memiliki boxplotnya tersendiri
# library yang diperlukan
import matplotlib.pyplot as plt
import seaborn as sms

# Membuat boxplot untuk fitur numerik masing-masing
numeric_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    df.boxplot(column=[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel('Values')

plt.tight_layout()
plt.show()
```

```python
# Promt : buat kode untuk membuat scater plot untuk semua fitur numerik
# library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 8))
for i, col1 in enumerate(numeric_columns):
    for j, col2 in enumerate(numeric_columns):
        if i != j:
            plt.subplot(len(numeric_columns), len(numeric_columns), i * len(numeric_columns) + j + 1)
            sns.scatterplot(data=df, x=col1, y=col2)
            plt.xlabel(col1)
            plt.ylabel(col2)

plt.tight_layout()
plt.show()
```

```python
# Promt : buat heat map dari kolerasi fitur numerik
# Menghitung korelasi antar fitur numerik
numeric_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
correlation_matrix = df[numeric_columns].corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Heatmap of Correlation between Numeric Features')
plt.show()
```
