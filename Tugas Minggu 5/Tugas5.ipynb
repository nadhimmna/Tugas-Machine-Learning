```python
# Promt : beri kode untuk memperbarui nama nama fitur
# Memperbarui nama kolom
df.columns = [
    'Timestamp',
    'Age',
    'Gender',
    'City',
    'Position',
    'Total_Experience',
    'Experience_in_Germany',
    'Seniority_Level',
    'Main_Technology',
    'Other_Technologies',
    'Yearly_Salary',
    'Yearly_Bonus_Stocks',
    'Previous_Year_Salary',
    'Previous_Year_Bonus_Stocks',
    'Vacation_Days',
    'Employment_Status',
    'Contract_Duration',
    'Main_Language',
    'Company_Size',
    'Company_Type',
    'Job_Loss_Due_to_Covid',
    'Reduced_Working_Hours',
    'Additional_Monetary_Support'
]

# Menampilkan data setelah pembaruan nama kolom
print("\nLima baris pertama dari data setelah pembaruan nama kolom:")
df.head()
```

```python
# Promt : beri kode untuk memperbarui nama nama fitur
# Menangani nilai yang hilang
print("\nJumlah nilai yang hilang untuk setiap kolom:")
print(df.isnull().sum())
```

```python
# Promt : beri kode untuk membersihkan dataset
# Membersihkan data
print("Status:")
# Menghapus baris yang memiliki nilai null
df.dropna(inplace=True)
# Menghapus duplikat baris
df.drop_duplicates(inplace=True)
# Mengonversi tipe data yang salah
df['Age'] = df['Age'].astype(int)
# Mengonversi format kolom 'Timestamp' menjadi datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
# Mengonversi kolom 'Yearly_Bonus_Stocks' dan 'Previous_Year_Bonus_Stocks' ke tipe data numerik
df['Yearly_Bonus_Stocks'] = pd.to_numeric(df['Yearly_Bonus_Stocks'], errors='coerce')
df['Previous_Year_Bonus_Stocks'] = pd.to_numeric(df['Previous_Year_Bonus_Stocks'], errors='coerce')
print("Data berhasil dibersihkan.")
```

```python
# Promt : beri kode untuk memvisualisasikan dataset
# library yang diperlukan
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca file CSV
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/IT Salary Survey EU  2020.csv')

# Pilih hanya kolom-kolom numerik untuk heatmap korelasi
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Buat subset dari dataframe yang hanya berisi kolom numerik
df_numeric = df[numeric_cols]

# Visualisasi Histogram untuk kolom 'Age'
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='Age', bins=20, kde=True, color='skyblue')
plt.title('Distribusi Usia')
plt.xlabel('Usia')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualisasi Heatmap untuk korelasi antara kolom numerik
plt.figure(figsize=(6, 4))
heatmap = sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
heatmap.set_title('Heatmap Korelasi antar Fitur', pad=20)
plt.tight_layout()
plt.show()
```
